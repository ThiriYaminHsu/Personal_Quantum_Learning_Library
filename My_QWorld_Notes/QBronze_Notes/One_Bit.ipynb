## **One Bit**

In daily life, we use the decimal (base-10) system for counting: 0-9. In computer science, the binary (base-2) system with only 0 and 1 is widely used. A bit (binary digit) is the basic unit of information and can be in either state 0 or 1.

**Why base-2?**

The fundamental architectures of most electronics are based on the flow of electricity. Just like we have ten fingers, electronic systems primarily have two states: on and off. When electricity is flowing, the state is denoted as 1, and when it is not flowing, it is in state 0.

### Four Operators of Classical Computing

1. **Identity (I)**
 * $ I(0) \rightarrow 0 $
 * $ I(1) \rightarrow 1 $

2. **Negation (NOT)**
 * $ NOT(0) \rightarrow 1 $
 * $ NOT(1) \rightarrow 0 $

3. **Constant (Zero)**
 * $ ZERO(0) \rightarrow 0 $
 * $ ZERO(1) \rightarrow 0 $

4. **Constant (One)**
 * $ ONE(0) \rightarrow 1 $
 * $ ONE(1) \rightarrow 1 $

### Reversibility and Irreversibility

* **Reversible Operators**: Identity and NOT, because the initial state can be determined from the final state.

* **Irreversible Operators**: ZERO and ONE, because the initial state cannot be determined from the final state.

This classification is crucial, as quantum evolution operators are reversible in closed systems.

The Identity and NOT operators are fundamental in quantum computing.
